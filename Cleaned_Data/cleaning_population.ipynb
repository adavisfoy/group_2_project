## Import Data

```python
# Read tornadoes csv into DataFrame
file_path = Path('../Raw_Data/Population_county_2010_2019_encoded.csv')
population_df = pd.read_csv(file_path)
population_df
```

```python
population_clean = population_df [['STATE', 'County FIPS', 'STNAME', 'CTYNAME', 
                                   'POPESTIMATE2010', 'POPESTIMATE2011', 'POPESTIMATE2012', 'POPESTIMATE2013', 
                                   'POPESTIMATE2014', 'POPESTIMATE2015', 'POPESTIMATE2016', 'POPESTIMATE2017', 
                                   'POPESTIMATE2018', 'POPESTIMATE2019']]
```
